> **EXPLORANDO VARIAVES CATEGORICA**
---

```python
plt.figure(figsize=(10, 5))

p = sns.countplot(data=df,x='churn')
plt.title("Clientes que deixaram de usar o serviço ou não" )
plt.xlabel("Churn")
plt.ylabel("Quantidade")

for valor in p.patches:
    altura = valor.get_height()

    plt.text(valor.get_x() + valor.get_width() / 2,
             altura + 0.3, 
             str(int(altura)),
             ha ='center')

plt.show()
```

```python
plt.figure(figsize=(10,6)) 

p = sns.countplot(data=df, x='pais', hue='churn')
plt.title("Distribuição de Churn por faixa de País")
plt.xlabel("País")
plt.ylabel("Quantidade")

for valor in p.patches:
    altura = valor.get_height()

    plt.text(valor.get_x() + valor.get_width() /2, 
            altura +0.5,
            str(int(altura)),
            ha='center')
    

plt.show()
```
